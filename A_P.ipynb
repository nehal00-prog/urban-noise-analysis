<h2 style="font-family:Times New Roman;color:Darkblue;">1.--- LINE PLOTS  ---</h2>

```python
# Group by hour and count car horn presence
car_horn_hourly = df.groupby('hour')['5-1_car-horn_presence'].sum()

# Plot the line chart
plt.figure(figsize=(10, 5))
plt.plot(car_horn_hourly.index, car_horn_hourly.values, marker='o', linestyle='-', color='tomato')
plt.title("Hourly Car Horn Frequency")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Car Horn Events")
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

<h2 style="color:lightgreen;">From the above gragh we observe that Frequency of car horn is maximum at 19 hour and minimum at 8 hour. </h2>

```python
# Define the noise tags to analyze
tags_to_plot = [
    '5-1_car-horn_presence',
    '5-3_siren_presence',
    '6-1_stationary-music_presence',
    '7-1_person-or-small-group-talking_presence'
]

# Group by hour and sum each tag's presence
tag_hourly_data = new_df.groupby('hour')[tags_to_plot].sum()
print(tag_hourly_data )
```

```python
# Plotting the grapg
plt.figure(figsize=(12, 6))
for tag in tags_to_plot:
    # Clean up label names for legend
    label = tag.replace('_presence', '').replace('-', ' ').replace('5 1', 'Car Horn').replace('5 3', 'Siren').replace('6 1', 'Music').replace('7 1', 'Voice')
    plt.plot(tag_hourly_data.index, tag_hourly_data[tag], marker='o', label=label)

plt.title("Hourly Trends of Urban Noise Events")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Events Detected")
plt.xticks(range(0, 24))
plt.legend(title="Noise Type")
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h2 style="color:lightgreen;">From the above gragh we observe that :-
    
- Car horns peak during rush hours (8 AM & 5‚Äì6 PM).

- Sirens are fairly evenly distributed, with small midday spikes.

- Music is most frequent in the afternoon and evening.

- Human voices peak during the daytime and drop late at night. </h2>
